```python
# Cell 1: Setup
"""
Boltzmann Distribution Lab - Interactive Demo
"""

from src.particle_system import ParticleSystem
from src.boltzmann_theory import BoltzmannTheory
from src.database import BoltzmannDatabase
from src.simulation_runner import SimulationRunner
from src.visualization import plot_speed_distribution
import numpy as np

# Cell 2: Single Simulation
"""
Run a single simulation and compare with theory
"""

# Parameters
n_particles = 10000
temperature = 300  # Room temperature
mass = 4.65e-26  # N2 molecule

# Create system
system = ParticleSystem(n_particles, temperature, mass)
theory = BoltzmannTheory(mass, temperature)

# Plot
speeds = system.get_speeds()
plot_speed_distribution(speeds, theory)

# Cell 3: Store in Database
"""
Store simulation in database
"""

with SimulationRunner() as runner:
    sim_id = runner.run_and_store(
        n_particles=n_particles,
        temperature=temperature,
        mass=mass,
        description="Room temperature N2"
    )
    print(f"Stored as simulation ID: {sim_id}")

# Cell 4: Query Database
"""
Query and retrieve data
"""

with BoltzmannDatabase() as db:
    # Get simulation
    sim = db.get_simulation_by_id(sim_id)
    print("Simulation metadata:")
    for key, value in sim.items():
        print(f"  {key}: {value}")
    
    # Get statistics
    stats = db.get_statistics(sim_id)
    print("\nStatistics:")
    print(f"  Mean speed: {stats['mean_speed']:.4e} m/s")
    print(f"  Theory prediction: {stats['mean_speed_theory']:.4e} m/s")

# Cell 5: Temperature Sweep
"""
Run multiple temperatures
"""

temperatures = np.linspace(100, 1000, 10)

with SimulationRunner() as runner:
    sim_ids = runner.run_temperature_sweep(
        temperatures=temperatures.tolist(),
        n_particles=5000,
        mass=mass
    )

# Cell 6: Visualize Results
"""
Visualize temperature sweep
"""

from src.database_visualization import (
    plot_temperature_sweep_speeds,
    plot_statistics_vs_temperature
)

with BoltzmannDatabase() as db:
    plot_temperature_sweep_speeds(db, sim_ids)
    plot_statistics_vs_temperature(db)
```
